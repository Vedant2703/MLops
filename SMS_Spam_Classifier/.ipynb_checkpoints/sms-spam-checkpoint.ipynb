# Data Cleaning

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
df
```

```python
df
```

```python
df.rename(columns={'v1':'Target','v2':'Text'},inplace=True)
df.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# fit_transform(y)
# Fit label encoder and return encoded labels

# Encode target labels with value between 0 and n_classes-1.
```

```python
df['Target']= encoder.fit_transform(df['Target'])
df

# ham=0
# spam=1
```

```python
# lets see if null value is present or not  
df.isnull().sum()
```

```python
# check for duplicate values 
df.duplicated().sum()
```

```python
# keep{‘first’, ‘last’, False}, default ‘first’
# Determines which duplicates (if any) to keep.
# ‘first’ : Drop duplicates except for the first occurrence.

df = df.drop_duplicates(keep='first')   
```

```python
df.duplicated().sum()
```

```python
nltk.download('punkt')
```

```python
df['Text'].apply(len)
```

```python
df['num_characters']= df['Text'].apply(len)
df
```

```python
df['Text'].apply(lambda x:nltk.word_tokenize(x))
```

```python
# count the number of words 
df['num_word'] = df['Text'].apply(lambda x:len(nltk.word_tokenize(x)))
df
```

```python
# count number of sentences
df['num_sentences'] = df['Text'].apply(lambda x:len(nltk.sent_tokenize(x)))
df
```

```python
df[['num_characters','num_word','num_sentences']].describe()
```

```python
df[df['Target']==0][['num_characters','num_word','num_sentences']].describe()
# ham
```

```python
df[df['Target']==1][['num_characters','num_word','num_sentences']].describe()
# spam
```
